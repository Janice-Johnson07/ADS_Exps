```python
#impuding values using mean
#for MASHTHICKNESS (amount of water added per pound of wheat)
missing_col = ['MashThickness']
```

```python
for i in missing_col:
  df.loc[df.loc[:,i].isnull(),i]=df.loc[:,i].mean()     
```

```python
#median method
df1 = pd.read_csv('/content/recipeData.csv', encoding='latin-1')
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
#impuding values using median
#for MASHTHICKNESS (amount of water added per pound of wheat)
missing_col_1 = ['MashThickness']
```

```python
for i in missing_col_1:
  df1.loc[df1.loc[:,i].isnull(),i]=df1.loc[:,i].median()
```

```python
df1.isnull().sum()
```

```python
#Filling with a new category
df1['UserId'] = df1['UserId'].fillna('No User ID')
df1['PitchRate'] = df1['PitchRate'].fillna('No PitchRate')
```
